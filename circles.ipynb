```python
# Program to plot a Circle
# using Parametric equation of a Circle
 
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['figure.facecolor'] = 'white'
plt.rcParams['figure.figsize'] = (20,20)




fig, ax = plt.subplots( 1 )
ax.set_facecolor('white')
ax.set_aspect( 1 )
ax.axis('off')
plt.xlim([-2, 2])
plt.ylim([-2, 2])

theta = np.linspace( 0 , 2 * np.pi-0.001 , 150 ) 

rmax=1
div = 50
colors = plt.cm.binary(np.linspace(0,1,div))

for i in range(1,3):
    radius = 1
    a = radius * np.cos( theta )
    b = radius * np.sin( theta )
    ax.plot( a, b, color=colors[0] )

for i in range(0,div):
    offx=i/2
    radius = np.sqrt(rmax**2 + offx**2)
    a = radius * np.cos( theta )
    a+= offx
    b = radius * np.sin( theta )
    ax.plot( a, b, color=colors[40],linewidth=1/radius )
    a = -a
    ax.plot( a, b, color=colors[40],linewidth=1/radius  )

for i in range(0,div):
    offx=i/2
    radius = np.sqrt(rmax**2 + offx**2)
    a = radius * np.cos( theta )
    b = radius * np.sin( theta )
    b+= offx
    ax.plot( a, b, color=colors[40],linewidth=1/radius )
    b = -b
    ax.plot( a, b, color=colors[40],linewidth=1/radius  )


    
plt.title( '' )
plt.show()
```
